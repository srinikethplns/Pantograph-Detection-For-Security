```python
import cv2
import numpy as np
from ultralytics import YOLO
import math

model_path = '/content/drive/MyDrive/Contactpoint.yolov8/runs/detect/train/weights/best.pt'
model = YOLO(model_path)

input_video_path = '/content/drive/MyDrive/panto5.mp4'
output_video_path = '/content/drive/MyDrive/processed_video1.mp4'

cap = cv2.VideoCapture(input_video_path)
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(cap.get(cv2.CAP_PROP_FPS))


fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter(output_video_path, fourcc, fps, (frame_width, frame_height))

frame_count = 0
initial_pantograph_coords = None

def find_contact_points(line, box):
    x1, y1, x2, y2 = box
    box_lines = [
        [(x1, y1), (x2, y1)],
        [(x2, y1), (x2, y2)],
        [(x2, y2), (x1, y2)],
        [(x1, y2), (x1, y1)]
    ]
    contact_points = []
    for box_line in box_lines:
        intersection = line_intersection(line, box_line)
        if intersection and x1 <= intersection[0] <= x2 and y1 <= intersection[1] <= y2:
            contact_points.append(intersection)
    return contact_points

def line_intersection(line1, line2):
    def line_eq(p1, p2):
        A = p1[1] - p2[1]
        B = p2[0] - p1[0]
        C = p1[0] * p2[1] - p2[0] * p1[1]
        return A, B, -C

    p1, p2 = line1
    q1, q2 = line2
    L1 = line_eq(p1, p2)
    L2 = line_eq(q1, q2)
    D = L1[0] * L2[1] - L1[1] * L2[0]
    Dx = L1[2] * L2[1] - L1[1] * L2[2]
    Dy = L1[0] * L2[2] - L1[2] * L2[0]
    if D != 0:
        x = Dx / D
        y = Dy / D
        return int(x), int(y)
    return None

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break


    results = model(frame)

    pantograph_box = None
    pantobar_box = None
    cable_boxes = []


    for result in results:
        for box in result.boxes:
            class_id = int(box.cls[0])
            x1, y1, x2, y2 = map(int, box.xyxy[0])

            if class_id == 3:
                pantograph_box = (x1, y1, x2, y2)
            elif class_id == 2:
                pantobar_box = (x1, y1, x2, y2)
            elif class_id == 1:
                cable_boxes.append((x1, y1, x2, y2))

    if pantograph_box is not None and pantobar_box is not None:
        if initial_pantograph_coords is None:
            initial_pantograph_coords = pantograph_box

        bx1, by1, bx2, by2 = pantobar_box


        cv2.rectangle(frame, (bx1, by1), (bx2, by2), (0, 255, 255), 2)


        pantobar_top_line = [(bx1, by1), (bx2, by1)]
        cv2.line(frame, pantobar_top_line[0], pantobar_top_line[1], (255, 0, 0), 2)
        blue_line_center = ((bx1 + bx2) // 2, by1)


        distances = []


        cable_boxes.sort(key=lambda box: box[0])


        for i, cable_box in enumerate(cable_boxes):
            cx1, cy1, cx2, cy2 = cable_box


            cv2.rectangle(frame, (cx1, cy1), (cx2, cy2), (0, 0, 255), 2)


            contact_points = find_contact_points(pantobar_top_line, cable_box)

            if len(contact_points) >= 2:

                midpoint = ((contact_points[0][0] + contact_points[1][0]) // 2,
                            (contact_points[0][1] + contact_points[1][1]) // 2)
                cv2.circle(frame, midpoint, 6, (255, 255, 255), -1)


                distance = math.sqrt((midpoint[0] - blue_line_center[0])**2 + (midpoint[1] - blue_line_center[1])**2)
                distance_meters = distance / (bx2 - bx1) * 3
                distances.append(distance_meters)


        for i, distance in enumerate(distances):
            label = f'Cable {i+1}: {distance:.4f} m'
            cv2.putText(frame, label, (10, 30 + 20*i), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 255, 0), 2)


        initial_x1, initial_y1, initial_x2, initial_y2 = initial_pantograph_coords
        current_x1, current_y1, current_x2, current_y2 = pantograph_box

        vertical_movement = (current_y1 + current_y2) / 2 - (initial_y1 + initial_y2) / 2
        horizontal_movement = (current_x1 + current_x2) / 2 - (initial_x1 + initial_x2) / 2


        vertical_movement_meters = vertical_movement / (by2 - by1) * 3
        horizontal_movement_meters = horizontal_movement / (bx2 - bx1) * 3


        cv2.putText(frame, f'Panto movt ver: {vertical_movement_meters:.2f} m', (10, frame_height - 30), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 255, 0), 2)
        cv2.putText(frame, f'Panto movt hor: {horizontal_movement_meters:.2f} m', (10, frame_height - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 255, 0), 2)

        print(f"Pantograph movement - Vertical: {vertical_movement_meters:.2f} m, Horizontal: {horizontal_movement_meters:.2f} m")
        print(f"Pantobar box: {pantobar_box}")
        print(f"Cable boxes: {cable_boxes}")


    out.write(frame)

    frame_count += 1

cap.release()
out.release()
cv2.destroyAllWindows()
```
